```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sem1 = pd.read_csv('user_fechas_W01.csv')
sem2 = pd.read_csv('user_fechas_W2.csv')
sem3 = pd.read_csv('user_fechas_W3.csv')
sem4 = pd.read_csv('user_fechas_W4.csv')
sem5 = pd.read_csv('user_fechas_W5.csv')
sem6 = pd.read_csv('user_fechas_W6.csv')
total = pd.read_csv('user_fechas.csv')

df_total = pd.DataFrame(total)
del df_total['Final']
corr_df_total = df_total.corr(method='pearson')

df_sem1 = pd.DataFrame(sem1)
del df_sem1['Final']
corr_df_sem1 = df_sem1.corr(method='pearson')

df_sem2 = pd.DataFrame(sem2)
del df_sem2['Final']
corr_df_sem2 = df_sem2.corr(method='pearson')

df_sem3 = pd.DataFrame(sem3)
del df_sem3['Final']
corr_df_sem3 = df_sem3.corr(method='pearson')

df_sem4 = pd.DataFrame(sem4)
del df_sem4['Final']
corr_df_sem4 = df_sem4.corr(method='pearson')

df_sem5 = pd.DataFrame(sem5)
del df_sem5['Final']
corr_df_sem5 = df_sem5.corr(method='pearson')

df_sem6 = pd.DataFrame(sem6)
del df_sem6['Final']
corr_df_sem6 = df_sem6.corr(method='pearson')


corr_table = pd.DataFrame()
corr_table['Week 1'] = corr_df_sem1['Nota']
#corr_table['Week 2'] = corr_df_sem2['Nota']
corr_table['Week 3'] = corr_df_sem3['Nota']
#corr_table['Week 4'] = corr_df_sem4['Nota']
corr_table['Week 5'] = corr_df_sem5['Nota']
#corr_table['sem6'] = corr_df_sem6['Nota']
corr_table['All']= corr_df_total['Nota']
corr = corr_table.drop(corr_table.index[[0]])

corr.index = ["Video_Video", "Video_Forum", "Video_Forum_Search",
                 "Video_Forum_Response", "Video_Forum_Comment",
                 "Video_Forum_Created", "Video_Problem", "Problem_Problem",
                 "Problem_Video", "Problem_Forum", "Problem_Forum_Search",
                 "Problem_Forum_Response", "Problem_Forum_Comment",
                 "Problem_Forum_Created", "Forum_Forum", "Forum_Video",
                 "Forum_Search_Video", "Forum_Response_Video",
                 "Forum_Comment_Video", "Forum_Created_Video", "Forum_Problem",
                 "Forum_Search_Problem", "Forum_Response_Problem",
                 "Forum_Comment_Problem", "Forum_Created_Problem",
                 "Forum_Comment_Forum_Comment", "Forum_Comment_Forum_Search",
                 "Forum_Comment_Forum_Response", "Forum_Comment_Forum_Created",
                 "Forum_Search_Forum_Comment", "Forum_Search_Forum_Search",
                 "Forum_Search_Forum_Response", "Forum_Search_Forum_Created",
                 "Forum_Response_Forum_Comment", "Forum_Response_Forum_Search", "Forum_Response_Forum_Response",
                 "Forum_Response_Forum_Created", "Forum_Created_Forum_Comment",
                 "Forum_Created_Forum_Search", "Forum_Created_Forum_Response",
                 "Forum_Created_Forum_Created","Grade"]
print(corr.head())
plt.figure(figsize=(3, 10))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.show()
```
